## KPI Feedback Report

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('emp_performance.csv')
data.head(5)
```

### Check primary key value is duplicate or not

```python
data['employee_id'].duplicated().sum()
```

### Check Data Infromation

```python
data.info()
```

### Dropped the unnecessary column

```python
data.drop(['no_of_trainings'], axis=1)
```

### Calculate the count of null value in each column

```python
data.isnull().sum()
```

### Calculate the mean value for Numeric column which contain null value

```python
age_mean = data['age'].mean().round()
previous_year_rating_mean = data['previous_year_rating'].mean().round()
current_rating_mean = data['current_rating'].mean().round()
performance_score_mean = data['performance_score'].mean().round()
```

### Filling `na` value with mean value in respective null value in column

```python
data['age'] = data['age'].fillna(age_mean).astype(int)
data['previous_year_rating'] = data['previous_year_rating'].fillna(previous_year_rating_mean).astype(int)
data['current_rating'] = data['current_rating'].fillna(current_rating_mean).astype(int)
data['performance_score'] = data['performance_score'].fillna(performance_score_mean).astype(int)
```

```python
data.info()
```

## Create a new column using dataset column

```python
data['is_promoted'] = data['performance_score'] > 70
```

```python
data.sample(5)
```

## Department wise promoted employee count

```python
gb = data.groupby(['department','is_promoted'])['employee_id'].count()
fig = gb.unstack(-1)
fig.plot.barh(figsize=(25,5))
```

## Department wise gender distribution

```python
plt.figure(figsize=(15,5))
fig = sns.countplot(data=data, x="department", hue="gender")
fig.bar_label(fig.containers[0])
plt.show()
```

## Employee strength % in each department

```python
plt.figure(figsize=(10,5))
key = data['department'].value_counts().keys()
plt.pie(data['department'].value_counts(),startangle=60,labels=key,shadow=True,autopct='%1.1f%%')
plt.show()
```

## Rating wise employee count

```python
data['current_rating'].value_counts().plot.barh()
plt.xlabel('count')
```

## Employee count under respective lead_name

```python
data['lead_name'].value_counts()
```

## Department wise education with employee count

```python
gb = data.groupby(['department','education'])['employee_id'].count()
fig = gb.unstack(-1)
fig.plot.bar(figsize=(25,5))
```

## Lead wise promoted employee count

```python
gb = data.groupby(['lead_name','is_promoted'])['employee_id'].count()
fig = gb.unstack(-1)
fig.plot.barh(figsize=(10,25))
```
